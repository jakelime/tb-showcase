```python
full_txt_str = ""

part_txt_str = f"""Hi VertexAI, I need you pretend that you are a 
production planner for my engines repair shop. We deal with MRO 
of jet engines.

These are the 4 data tables that you need.

Table1: A list of jobs and their expected start and end times.

{df_job_details.to_csv()}

In table1, you can find the priority of each job by the expected 
end time.


Table2: A table of tasks and their number of hours required to 
complete them.

{df_workscope.to_csv()}

The "task" column is the task and the "manhour_to_be_planned" 
is the number of manhours required to perform that task.

The "task" string in each row can be decomposed in this 
template "EngineMajorModule-EngineSubModule-TaskSequence-RepairName".

To set the precedence for the tasks: the engine has to be stripped into major
modules. There are 5 major modules:
['lpt_major_module-*-*-*', 'accessory_drive_module-*-*-*', 'engine_major-*-*-*', 
'fan_major_module-*-*-*', 'core_major_module-*-*-*', 'others-*-*-*']. Each of 
these major modules are independent of each other and can be done in parallel.

Within each major module, it can be further stripped into sub-modules.
Different sub-module can be done in parallel, but the tasks within the sub-module 
are done in sequence.
For example, the sub-module 'lpt_major_module-CUCEAA' must be done in this order: 
['lpt_major_module-CUCEAA-010-*', 'lpt_major_module-CUCEAA-021-*',
 'lpt_major_module-CUCEAA-022-*', 'lpt_major_module-CUCEAA-030-*']


Table3: A resource allocation table to list all possible tasks, 
the machines and the technicians required to perform them.

{df_shopfloor.to_csv()}

"machine_ids" is a list of machines that can be used to perform the task, separated
by semicommas.

"emp_ids" is a list of employees that can be used to perform the task, separated
by semicommas.

Table3: A manpower table to list all technicians, their work shift, when is the start
of their shift, and when is the end of their shift and which day of the week they are
working.

{df_manpower.to_csv()}


The primary goal is to minimize the total time taken to complete all jobs, prioritising
jobs with earlier due dates. You will need to efficiently minimize idle time for machines and 
techinicians.

I need the planner to perform the order of tasks, generate a schedule in CSV, with the c. 

I need the schedule in this table with columns [\"job_id\", \"task\", \"machine_id\",
\"emp_id\", \"start_datetime\", \"end_datetime\"]. Can you reply back in this chat, in CSV format?
"""

full_txt_str = f"{full_txt_str}\n\n{part_txt_str}\n\n"
```

```python
ai = AiModel()
res = ai.generate(full_txt_str)
ai.get_responses()
```
